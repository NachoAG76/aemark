# 1. Módulo Matplotlib

La librería __matplotlib__ ofrece métodos para la visualización de datos. 
* Se trata de la librería gráfica de Python. 
* La  mayor parte de las funciones gráficas se encuentran en la sublibrería __pyplot__.

```python
import numpy as np                         # importación de numpy
import matplotlib as mpl                   # importación de matplotlib con alias mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1.1 Líneas. Función `plot`

Lo primero que vamos a hacer es generar los datos para después representarlos.

```python
# generación de datos
x = np.linspace(start=0, stop=2*np.pi, num=50)
y1 = np.sin(x)
y2 = np.cos(x)                  
```

La función __plot__ representa los datos contenidos en __x__ e __y__. Como podemos ver en la figura siguiente, se trata de una gráfica muy simple y con muy poca información. 
A partir de ella, iremos añadiendo más información, como el título de la gráfica, la etiqueta de cada uno de los ejes, la leyenda en caso necesario, etc.


```python
fig, ax = plt.subplots(1, figsize=(12,6))
ax.plot(x, y1, label='seno')
ax.plot(x, y2, label='coseno')

ax.legend(fontsize=16)
ax.axes.xaxis.set_label_text("X Axis", fontdict={"size":15})
ax.axes.yaxis.set_label_text("Y Axis", fontdict={"size":15})
ax.title.set_text("Gráficos con Matplotlib")

```

### 1.1.1 Opciones de estilo

El estilo por defecto de las gráficas es muy simple, pero __matplotlib__ ofrece un gran número de opciones que permiten cambiar y enriquecer el estilo.

```python
fig, ax = plt.subplots(1, figsize=(12,6))
ax.grid(True)            # Opción: Pinta una cuadrícula en la gráfica

ax.plot(x, y1, label='seno')
ax.plot(x, y2, label='coseno')

ax.legend(fontsize=16)
ax.axes.xaxis.set_label_text("X Axis", fontdict={"size":15})
ax.axes.yaxis.set_label_text("Y Axis", fontdict={"size":15})
ax.title.set_text("Gráficos con Matplotlib")
ax.legend(loc=3);        # loc=0: El mejor sitio para colocar la leyenda
```

__Cambiando el estilo de representación__

Por defecto la función __plot__ representa los datos mediante una línea continua de color azul. Si queremos representar los datos discretos mediante puntos, tan solo tenemos que cambiar las opciones de estilo. 

Fíjate en la primera línea de código de la siguiente celda:

```python

fig, ax = plt.subplots(1, figsize=(12,6))
ax.grid(True)            # Opción: Pinta una cuadrícula en la gráfica

# pinta los datos con puntos (opción: 'o') rojos(opción:'r')
ax.plot(x, y1,'ro', label='seno')    
ax.plot(x, y2, 'b--', label='coseno')

ax.legend(fontsize=16)
ax.axes.xaxis.set_label_text("X Axis", fontdict={"size":15})
ax.axes.yaxis.set_label_text("Y Axis", fontdict={"size":15})
ax.title.set_text("Gráficos con Matplotlib")
ax.legend(loc=0);        # loc=0: El mejor sitio para colocar la leyenda
```

* Los colores y estilos se pueden combinar de cualquier forma para distinguir diferentes conjuntos de datos.

## 1.2 Histogramas. Función `hist`

Permiten representar de forma gráfica una variable en forma de barras, donde cada barra es proporcional a la frecuencia de los datos representados. Creamos este tipo de gráficos con la función __hist__.

```python
datos = np.random.randn(100) * 100.0
media = datos.mean()              # media
desviacion = datos.std()          # desviación
m_y = 1.5

fig, ax = plt.subplots(1, figsize=(12,6))
ax.hist(datos, bins = 15)   # bin = número de barras a representar
ax.plot(media, m_y, "ko")
ax.plot([media, media] , [0, 12], "r"); # pinta una línea --

ax.set_title('Histograma')
ax.axes.xaxis.set_label_text("Valores", fontdict={"size":15})
ax.axes.yaxis.set_label_text("Frecuencia", fontdict={"size":15});
```

La función __plt.hist__ soporta muchas opciones de configuración. El parámetro __bins__ en la función permite indicar el número de barras a representar. 

En el siguiente ejemplo optamos por representar las dos variables en la misma figura, cada una de ellas con su etiqueta (__label__):

```python
datos = np.random.randn(100, 2) * 100.0     # 100 filas y 2 columnas

# Creamos una figura 
fig, ax = plt.subplots(1, figsize=(12,6))

# representamos las dos variables en el mismo histograma
ax.hist(datos, label=['columna0', 'columna 1'], bins = 25)  
ax.legend(loc=0)

ax.set_title('Histograma');
```

## 1.3 Diagramas de caja. Función `boxplot`

Este tipo de diagramas se construyen mediante la función __boxplot__.

```python
# Creamos una figura 
fig, ax = plt.subplots(1, figsize=(12,6))

plt.figure(figsize=(7,3))
ax.boxplot(datos)

ax.grid(True)
ax.set_title('Diagrama de caja o Bigotes')

ax.set_xticks([1, 2],[ 'Columna 1', 'Columna 2']);


fig.savefig('mi_boxplot1.pdf')     # para guardar los gráficos
fig.savefig('mi_boxplot2.png');
```

La función __plt.savefig__ permite guardar los gráficos en diferentes formatos (pdf, png, ps, etc).

## 1.4 Diagramas de dispersión. Función `scatter`

Se usa habitualmente para observar la relación entre dos variables.
Vemos un ejemplo:

```python
# nos creamos datos de forma aleatoria:
datos = np.random.standard_normal((500, 2))   # 500 filas y 2 columnas (2 variables)
datos = datos.cumsum(axis = 0)

datos[:10]         # mostramos las 10 primeras filas
```

Creamos un digrama de dispersión con la función __scatter__.

```python
variable1 = datos[:,0]
variable2 = datos[:,1]

fig, ax = plt.subplots(1, figsize=(12,6))
ax.set_title('Diagrama de dispersion')
ax.scatter(variable1, variable2, marker = 'o')    # puntitos azules

plt.xlabel('Variable 1')
plt.ylabel('Variable 2');

ax.grid(True)
```

## 1.5 Subgráficos

Permite representar información en una misma figura.

```python
# generamos datos aleatorios
x = np.arange(0, 100)
y = np.random.rand(100)  # 100 random numbers

# generamos 4 sub-gráficos
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2,figsize=(7,5))  
ax1.plot(x, y)
ax2.hist(y)
ax3.scatter(x, y)
ax4.boxplot(y);
```
